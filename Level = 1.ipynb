```python
X=df.iloc[:,:-1]
Y=df.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Perform PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # Set the number of components you want to keep
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
X_train_pca.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

cls = RandomForestClassifier()
cls.fit(X_train_pca, y_train)
```

```python


# Predict the target variable for the test set
y_pred = cls.predict(X_test_pca)

from sklearn.metrics import accuracy_score

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# Load the dataset (example with Iris dataset)
data = load_iris().data

# Step 1: Standardization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Step 2: PCA
pca = PCA(n_components=2)  # Specify the number of components you want to retain
principal_components = pca.fit_transform(scaled_data)

# Accessing the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Line plot using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Components')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.show()
```
